```python
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings("ignore")
```

```python
XOM = pd.read_csv('XOM.csv')
```

```python
XOM.head()
```

```python
XOM2 = XOM[XOM['Year'].isin([2019,2020])]
```

```python
np.max(XOM[XOM['Year'].isin([2019])].Week_Number)
```

```python
np.max(XOM[XOM['Year'].isin([2020])].Week_Number)
```

```python
week_sign = []
for i in [2019, 2020]:
    for j in range(53):
        subset = XOM2[(XOM2.Week_Number == j) & (XOM.Year == i)]
        n = subset.shape[0]
        sign = np.where(np.mean(subset.Return >= 0) >= 0.5, '+', '-')
        for k in range(n):
            week_sign.append(sign.tolist())
```

```python
XOM2['week_sign'] = week_sign
```
